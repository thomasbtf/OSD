```python
if not debug:
    osd_meta_data = pd.read_csv(snakemake.input[0], sep = "|")
else:
    osd_meta_data = pd.read_csv("/local/thomas/OSD/resources/OSD2014-env_data_2017-11-23.csv", sep = "|")
    display(osd_meta_data.head())
    display(osd_meta_data.columns)
```

```python
import altair as alt
from vega_datasets import data

sphere = alt.Chart(alt.sphere()).mark_geoshape(fill='lightblue')
graticule = alt.Chart(alt.graticule()).mark_geoshape(stroke='white', strokeWidth=0.5)

countries = alt.Chart(
    alt.topo_feature(data.world_110m.url, 'countries')
).mark_geoshape(
    stroke='grey',
    strokeWidth=1
).encode(
    color=alt.value('#eee'),
).properties(
    width=1000,
    height=500
).project(
    'naturalEarth1'
).properties(
    title="OSD 2014 Sites"
)

sample_points = alt.Chart(osd_meta_data).mark_point().encode(
    latitude='start_lat',
    longitude='start_lon',
    color=alt.Color('water_depth', scale=alt.Scale(scheme='purpleblue'), title="Water Depth", sort="descending"),
    opacity=alt.value(0.5),
    tooltip=["osd_id", "water_depth", "device", "description", "longhurst_descr"]
)

location_plot = (sphere + graticule + countries + sample_points).configure_view(stroke=None)

if not debug:
    (location_plot).save(snakemake.output[0])
else:   
    display(location_plot)
```
