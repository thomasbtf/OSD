```python
if "snakemake" in locals():
    debug = False
else:
    debug = True

if not debug:
    import sys
    sys.stderr = open(snakemake.log[0], "w")

import altair as alt
import pandas as pd
```

```python
if not debug:
    arg_call = pd.read_csv(snakemake.input[0], sep = "\t")
    sample = snakemake.params.get("sample", "")
else:
    arg_call = pd.read_csv("/local/thomas/OSD/results/deeparg/OSD1_ME_shotgun_workable_merged/OSD1_ME_shotgun_workable_merged.mapping.ARG", sep = "\t")
    sample = "OSD1"
    display(arg_call.head())
```

```python
source = arg_call["predicted_ARG-class"].value_counts().rename_axis('ARG-class').reset_index(name='counts')

bars = alt.Chart(source).mark_bar(color="#003f5c").encode(
    x=alt.X('counts:Q', title=f'Occurrences in OSD sample {sample}'),
    y=alt.Y('ARG-class:O', title='ARG-Class', sort='y')
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text=alt.Text('counts:Q', format=",.0f")
)

if not debug:
    (bars + text).save(snakemake.output[0])
else:   
    display((bars + text))
```
